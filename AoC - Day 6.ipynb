## Part 1 - Problem Statement

Except for the universal Center of Mass (COM), every object in space is in orbit around exactly one other object.

Before you use your map data to plot a course, you need to make sure it wasn't corrupted during the download. To verify maps, the Universal Orbit Map facility uses orbit count checksums - the total number of direct orbits (like the one shown above) and indirect orbits.

Whenever A orbits B and B orbits C, then A indirectly orbits C. This chain can be any number of objects long: if A orbits B, B orbits C, and C orbits D, then A indirectly orbits D.

For example, suppose you have the following map:

* COM)B
* B)C
* C)D
* D)E
* E)F
* B)G
* G)H
* D)I
* E)J
* J)K
* K)L

Visually, the above map of orbits looks like this:

        G - H       J - K - L
       /           /
COM - B - C - D - E - F

               \
                I
                
                
In this visual representation, when two objects are connected by a line, the one on the right directly orbits the one on the left.

Here, we can count the total number of orbits as follows:

* D directly orbits C and indirectly orbits B and COM, a total of 3 orbits.
* L directly orbits K and indirectly orbits J, E, D, C, B, and COM, a total of 7 orbits.
* COM orbits nothing.

The total number of direct and indirect orbits in this example is 42.

What is the total number of direct and indirect orbits in your map data?

## Solution

```python
from collections import defaultdict, deque
```

```python
def read_input(data):
    return tuple([dict(zip(('tail', 'head'), tuple(edge.split(')')))) for edge in data.split()])
```

```python
def build_graph(edges):
    tree = defaultdict(list)
    
    for edge in edges:
        tree[edge['tail']].append(edge['head'])
        
    return dict(tree)
```

```python
def bfs_distance(tree, root):
    distances = {root: 0}
    frontier = deque([root])
    while frontier:
        current_node = frontier.popleft()
        if current_node in tree:
            frontier.extend(tree[current_node])
            for child in tree[current_node]:
                distances[child] = distances[current_node] + 1
            
    return distances
```

## Tests

```python
assert read_input('COM)B B)C C)D') == ({'tail': 'COM', 'head': 'B'},
                                       {'tail': 'B', 'head': 'C'},
                                       {'tail': 'C', 'head': 'D'})
```

```python
test_data = 'COM)B B)C C)D D)E E)F B)G G)H D)I E)J J)K K)L'
assert build_graph(read_input(test_data)) == {
    'COM': ['B'],
    'B': ['C', 'G'],
    'C': ['D'],
    'D': ['E', 'I'],
    'E': ['F', 'J'],
    'G': ['H'],
    'J': ['K'],
    'K': ['L']}
```

```python
assert (bfs_distance(build_graph(read_input(test_data)), 'COM') == {
    'COM': 0,
    'B': 1,
    'C': 2,
    'G': 2,
    'D': 3,
    'H': 3,
    'E': 4,
    'I': 4,
    'F': 5,
    'J': 5,
    'K': 6,
    'L': 7  
})
```

```python
assert sum(bfs_distance(build_graph(read_input(test_data)), 'COM').values()) == 42
```

## Part 2 - Problem Statement

Now, you just need to figure out how many orbital transfers you (YOU) need to take to get to Santa (SAN).

You start at the object YOU are orbiting; your destination is the object SAN is orbiting. An orbital transfer lets you move from any object to an object orbiting or orbited by that object.

For example, suppose you have the following map:

* COM)B
* B)C
* C)D
* D)E
* E)F
* B)G
* G)H
* D)I
* E)J
* J)K
* K)L
* K)YOU
* I)SAN

Visually, the above map of orbits looks like this:

                          YOU
                         /
        G - H       J - K - L
       /           /
COM - B - C - D - E - F

               \
                I - SAN
                

In this example, YOU are in orbit around K, and SAN is in orbit around I. To move from K to I, a minimum of 4 orbital transfers are required:

* K to J
* J to E
* E to D
* D to I

What is the minimum number of orbital transfers required to move from the object YOU are orbiting to the object SAN is orbiting? (Between the objects they are orbiting - not between YOU and SAN.)

## Solution

```python
def build_reverse_graph(edges):
    tree = defaultdict(list)
    
    for edge in edges:
        tree[edge['head']] = edge['tail']
        
    return dict(tree)
```

```python
def find_path_to_root(reverse_tree, root, node):
    path = [node]
    
    while path[-1] != root:
        path.append(reverse_tree[path[-1]])
        
    return path
```

```python
orbit_data = read_input(data)
tree = build_graph(orbit_data)
distances = bfs_distance(tree, 'COM')

reverse_tree = build_reverse_graph(orbit_data)
you_path = find_path_to_root(reverse_tree, 'COM', 'YOU')
santa_path = find_path_to_root(reverse_tree, 'COM', 'SAN')
common_path = set(you_path).intersection(set(santa_path))

latest_common_ancestor = max(common_path, key=lambda node: distances[node])

((distances['YOU'] - distances[latest_common_ancestor]) +
 (distances['SAN'] - distances[latest_common_ancestor]) - 2)
```
